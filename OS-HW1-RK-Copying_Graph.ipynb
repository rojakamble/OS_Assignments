### Generate a graph that depicts the program performance. 
Provide your graph in a file named copying.jpg  

```python
# Giving file sizes and performance time(ran in particular machine) as inputs respectively

file = [0.001, 0.002, 0.004, 0.008, 0.016, 0.032, 0.064, 0.128, 0.256, 0.512, 1, 2, 4, 8, 16, 32, 64, 128]
real = [3, 2, 0 , 0 , 1, 1, 1, 1, 0, 1, 1, 1, 1, 0, 1, 1, 1, 1]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(30,10))
plt.plot(file, real, marker='o', color='green', markerfacecolor='red', label = ('File size', 'Time'))
plt.title("Graph of file size vs Time to copy", fontsize=25)
plt.xlabel('File size in (MB)', fontsize=25)
plt.ylabel('Time in (microseconds)', fontsize=25)
plt.legend()
plt.show()
plt.savefig('copying.png')
```
